```python

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
%reload_ext autoreload
%autoreload 2
import scipy.stats as st
import os
from scipy.optimize import curve_fit
import igraph as ig
import csv

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv('/Users/YuchenZheng/Downloads/year_origin_destination_hs07_4.tsv',sep='\t')
```

```python
df = df.fillna(0)
```

```python
mean = df.groupby(['origin','dest','hs07','year']).mean().reset_index()
```

```python
#filter out the xx "countries" (represent groups?)
filtered = mean[(~mean['origin'].str.contains("xx") & (~mean['dest'].str.contains("xx")))]
```

```python
pairs2008 = np.array(filtered[filtered['year']==2008][['origin','dest','hs07','export_val']])
```

```python
with open("/Users/YuchenZheng/Downloads/oridest2008.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(pairs2008)
```

```python
reader = csv.reader(open("/Users/YuchenZheng/Downloads/oridest2008.csv"))
graph2008 = ig.Graph.TupleList(reader, directed = True, weights=True)

```

```python
graph2008.vs['label'] = graph2008.vs['name']
```

```python
df2008 = pd.DataFrame()
df2008['country'] = graph2008.vs['name']
df2008['degree'] = graph2008.degree()
df2008['closeness'] = graph2008.closeness()
df2008['betweenness'] = graph2008.betweenness()
df2008['pagerank'] = graph2008.pagerank()
df2008['deg. assortativity'] = graph2008.assortativity_degree()
df2008['local clustering coefficient'] = graph2008.transitivity_local_undirected(mode ='zero')
df2008['global clustering coefficient'] = graph2008.transitivity_undirected()

```
